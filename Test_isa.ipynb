```python
# libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurações visuais
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# Carregando o CSV
df = pd.read_csv(r"C:\Users\bebel\anaconda3\br_sp_gov_ssp_ocorrencias_registradas.csv")
df.tail(3)
```

```python
# Informações gerais
df.info()
```

```python
# Verificando valores nulos
df.isnull().sum()
```

```python
df.fillna({
    'homicidio_doloso': 0,
    'numero_de_vitimas_em_homicidio_doloso': 0,
    'homicidio_culposo_por_acidente_de_transito': 0,
    'homicidio_culposo_outros': 0,
    'tentativa_de_homicidio': 0,
    'lesao_corporal_dolosa': 0,
    'lesao_corporal_culposa_por_acidente_de_transito': 0,
    'lesao_corporal_culposa_outras': 0,
    'latrocinio': 0,
    'numero_de_vitimas_em_latrocinio': 0,
    'total_de_estupro': 0,
    'roubo_de_veiculo': 0,
    'roubo_a_banco': 0,
    'roubo_de_carga': 0,
    'furto_outros': 0,
    'furto_de_veiculo': 0
}, inplace=True)
df.head(3)
```

```python
# Estatísticas básicas
df.describe()

```

```python
# - Muitos valores zero: A mediana (50%) de quase todos os crimes é zero — isso indica que em muitos municípios/meses não há registro de certos crimes.
# - Desvio padrão alto: Crimes como roubo e furto têm grande variação, o que pode indicar concentração em áreas urbanas.
```

```python
# Criando coluna de data completa
df['data'] = pd.to_datetime(df['ano'].astype(str) + '-' + df['mes'].astype(str) + '-01')
df.head(4)
```

```python
# Ordenando por data
df = df.sort_values('data')
df.head(3)

```

```python
# Verificando municípios únicos
df['id_municipio'].nunique()
df.head(3)
```

```python
df_grouped = df.groupby('data')[['homicidio_doloso', 'latrocinio', 'total_de_estupro']].sum()
df_grouped.plot(title='Evolução de Crimes Violentos em SP')
plt.xlabel("Data")
plt.ylabel("Número de Ocorrências")
plt.show()
```

```python
df_roubo = df.groupby('regiao_ssp')['total_de_roubo_outros'].sum().sort_values(ascending=False)

sns.barplot(x=df_roubo.values, y=df_roubo.index)
plt.title("Total de Roubos por Região SSP")
plt.xlabel("Ocorrências")
plt.ylabel("Região")
plt.show()
```

```python
df_crimes_regiao = df.groupby('regiao_ssp')[['roubo_outros', 'furto_outros', 'latrocinio']].sum()

df_crimes_regiao.plot(kind='bar', stacked=True, title='Distribuição de Crimes Patrimoniais por Região SSP')
plt.ylabel("Número de Ocorrências")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df[['homicidio_doloso', 'latrocinio', 'total_de_estupro', 'roubo_outros']])
plt.title("Distribuição dos Crimes com Outliers")
plt.show()
```

```python
# - a caixa é mais larga, os whiskers são mais altos e há muitos outliers.
# - Isso indica que roubos são muito mais frequentes e variam bastante entre regiões.
# - Os outliers mostram que alguns municípios têm números extremamente altos — esses são os que merecem atenção estratégica.
```

```python
# Definindo um limite para considerar como outlier
limite_outlier = df['roubo_outros'].quantile(0.99)  # top 1% mais altos

# Filtrando os registros acima do limite
df_outliers_roubo = df[df['roubo_outros'] > limite_outlier]

# Visualizando os municípios mais críticos
top_municipios_roubo = df_outliers_roubo.groupby('regiao_ssp')['roubo_outros'].sum().sort_values(ascending=False).head(10)

top_municipios_roubo.plot(kind='bar', title='Municípios com Mais Roubos (Outliers)')
plt.ylabel("Ocorrências de Roubo")
plt.xticks(rotation=45)
plt.show()
```

```python
df['indice_risco'] = (
    df['homicidio_doloso'] * 3 +
    df['latrocinio'] * 2 +
    df['roubo_outros'] +
    df['total_de_estupro'] * 2
)

# Agrupando por município
df_risco_municipio = df.groupby('regiao_ssp')['indice_risco'].sum().sort_values(ascending=False).head(10)

df_risco_municipio.plot(kind='bar', title='Top 10 Municípios com Maior Índice de Risco')
plt.ylabel("Índice de Risco")
plt.xticks(rotation=60)
plt.show()
```
